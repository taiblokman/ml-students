```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('student-mat.csv', sep=';')

# understand and visualize the correlation
plt.figure(figsize=(11,11))
sns.heatmap(df.corr().round(1), annot=True)

# choose the features more closely related to the final grade G3
df = df[['failures','G1','G2','G3']]
```

```python
df
```

### split the data into features vs target and create numpy arrays

```python
X = df.iloc[:, 0:3].values
y = df.iloc[:, -1].values
```

### split the data into training and testing
### and then create the linear regression model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
print(reg.score(X_test, y_test))
```

```python
y_pred = reg.predict(X_test)
plt.scatter(y_test, y_pred)
```
